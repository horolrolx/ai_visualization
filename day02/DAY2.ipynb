# PIL 라이브러리 예제

```python
from PIL import Image

img = Image.open("./coffee.jpg")
img
```

이미지 확인

```python
# 이미지 사이즈
print(img.size)
img.width, img.height
```

```python
# 이미지 포맷
img.format
```

```python
# 이미지 모드
img.mode
```

이미지 조작

```python
# 이미지 크기 변경
img_resized = img.resize((300,200))
img_resized
```

```python
# 이미지 회전
img_rotated = img_resized.rotate(90, expand=True)
img_rotated
```

```python
# 이미지 자르기
img_resized.crop((110,10,270,170))
```

```python
# 이미지 위에 박스 그리기
from PIL import ImageDraw

draw = ImageDraw.Draw(img_resized)
draw.rectangle((110,10,270,170), outline="red")
img_resized
```

```python
# 상하반전
img_resized.transpose(Image.FLIP_TOP_BOTTOM)
```

```python
# 색상변환
gray_img = img_resized.convert("L")
gray_img
gray_img.getbands()
```

```python
# 색상분리
r, g, b = img_resized.split()
Image.merge("RGB", (b, g, r))
```

```python
# 블러링
from PIL import ImageFilter
blur_img = img_resized.filter(ImageFilter.BLUR)
blur_img
```

```python
img_resized.filter(ImageFilter.BoxBlur(10))
```

```python
img_resized.filter(ImageFilter.SHARPEN)
```

```python
img_resized.filter(ImageFilter.SMOOTH)
```

```python
img_resized.filter(ImageFilter.FIND_EDGES)
```

```python
img_resized.filter(ImageFilter.EDGE_ENHANCE)
```

```python
img_resized.filter(ImageFilter.EMBOSS)
```

```python
from PIL import ImageEnhance
enhancer = ImageEnhance.Brightness(img_resized)
enhancer.enhance(2)
```
