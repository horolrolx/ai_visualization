```python
# LAB1
#   ,      .   BGR    
# img[100, 300] = [0, 0, 0]
import cv2
import matplotlib.pyplot as plt
img = cv2.imread('test_img.jpg', cv2.IMREAD_GRAYSCALE)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img[200, 300] = [0, 0, 0]
plt.imshow(img)
plt.show()

#  Numpy     numpy   .   
# roi_img = img[150:300, 50: 200]
roi_face = img[:320, 30:320]
plt.imshow(roi_face)
plt.show()
```
