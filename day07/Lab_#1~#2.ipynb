```python
# LAB1
# 픽셀값 읽기는 가로, 세로값을 입력하면 해당 위치의 픽셀값을 출력해준다. 해당 위치에 BGR 값을 대입하여 픽셀값을 변경하시오
# img[100, 300] = [0, 0, 0]
import cv2
import matplotlib.pyplot as plt
img = cv2.imread('test_img.jpg', cv2.IMREAD_GRAYSCALE)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img[200, 300] = [0, 0, 0]
plt.imshow(img)
plt.show()

# 이미지는 Numpy 배열로 이루어져 있기 때문에 numpy 함수를 사용해야 한다. 이미지에서 얼굴영역만 자르시오
# roi_img = img[150:300, 50: 200]
roi_face = img[:320, 30:320]
plt.imshow(roi_face)
plt.show()
```

```python
# LAB 2
# 이미지의 얼굴 영역에 찾아서 사각형과 텍스트를 출력하시오
# img = cv2.rectangle(img, (50, 150), (200, 300), (0,0,255), 3)
# img = cv2.putText(img, 'Cat', (50, 140),
# cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255,0,0), 2)

img_face_trace = cv2.rectangle(img, (30, 320), (320, 5), (0,0,255), 3)
plt.imshow(img_face_trace)
plt.show()

img_put_text = cv2.putText(img_face_trace, 'VELA MODEL', (100, 400), cv2.FONT_HERSHEY_SIMPLEX, 2, (255,0,0), 2)
plt.imshow(img_put_text)
plt.show()
```
